# Ajuste Cinetica de Oxidaci√≥n

```python
import pandas as pd
import numpy as np
from scipy.integrate import cumtrapz
from scipy.integrate import solve_ivp, odeint
from lmfit import minimize, Parameters, Parameter, report_fit
import matplotlib.pyplot as plt
plt.style.use('seaborn-notebook')

experimentos=['ON25','ON40','ON80','Aire25','Aire40','Aire80','N25','N40','N80'];
datos=[];
for hoja in experimentos:
    datos.append(pd.read_excel (r'../../Resultados Experimentales/TGA/Resumen_Datos.xlsx', sheet_name= hoja).to_numpy())
```

```python
ajustes=[]
for prueba in datos:
    ajustes.append(np.polyfit(prueba[:,0],prueba[:,1],10))
```

```python
def ODE_ajuste(C,t,ps):
    k=np.asarray([ps['k1'].value,ps['k2'].value,ps['k3'].value,ps['k4'].value,ps['k5'].value,ps['k6'].value]);
    dC=np.empty(len(C))
    dC[0]=-2*k[0]*C[0]**2+k[2]*C[1]*C[3];
    dC[1]=k[0]*C[0]**2+k[1]*C[2]*C[4]-k[2]*C[1]*C[3]-k[4]*C[2]*C[1]-2*k[5]*C[1]**2;
    dC[2]=k[0]*C[0]**2-k[1]*C[2]*C[4]+k[2]*C[1]*C[3]-2*k[3]*C[2]**2-k[4]*C[2]*C[1];
    dC[3]=-k[2]*C[1]*C[3];
    dC[-1]=0;
    
    return dC

def masa(ts,C,ps):
    sol = odeint(ODE_ajuste,C,ts,args=(ps,),rtol=1e-9,atol=1e-11);
    O2=(ps['k2'].value)*sol[:,-1]*sol[:,2]
    O2_int=(0.032/931)*cumtrapz(O2, ts, initial=0)
    return O2_int
    

def residual(ps, ts, data):
    x0 = ps['x0'].value, ps['y0'].value
    C=np.asarray([x0[0],0,0,x0[1],15640.8/(sol_ac)])
    model = masa(ts,C, ps)
    return (model - data).ravel()

# Solubilidad del oxigneo en aceite
sol_ac=1.58E+03 #[Pa m3/mol]

#Datos Experimentales
t=np.linspace(0,1440*60,1440*60+1);
funcion_1=np.poly1d(ajustes[5])
funcion_2=np.poly1d(ajustes[-1])
data=(funcion_1(t/60)-funcion_2(t/60))/100;


# set de parametros a ajustar
params = Parameters()
params.add('x0', value=0.005*1000,vary=1, min=0)
params.add('y0', value=6.8*1000,vary=1, min=0)
params.add('k1', value=6.60316288e-8)
params.add('k2', value=1e+04,vary=1)
params.add('k3', value=1.29581075e-02, vary=1, min=0)
params.add('k4', value=3e8, vary=1, min=0)
params.add('k5', value=1e7,vary=1,min=0)
params.add('k6', value=1e2, vary=1, min=0)
params=resultados_O2_80.params
# fit model and find predicted values
result = minimize(residual, params, args=(t, data), method='leastsq')
final = data + result.residual.reshape(data.shape)

# plot data and fitted curves
plt.plot(t, data, '-')
plt.plot(t, final, '-', linewidth=2);

# display fitted statistics
report_fit(result)
```

```python
ajuste_aire_80=[];
resultados_aire_80=result;
for a in list(result.params.valuesdict().items())[:]:ajuste_aire_80.append(a[1])
```

## Simulacion Oxidac√≠on en horno 

```python
#Ecuacion Diferencial a solucionar
D_ac=0.4e-9 #[m2/s]
sol_ac=1.58E+03 #[Pa m3/mol]

#Vector de radio discretizado
n=100
r=np.linspace(0,1.25e-4,n)
delta_r=r[1];

#Vector de inicializacion
ps=resultados_O2_80.params
x0 = resultados_O2_80.params['x0'].value, resultados_O2_80.params['y0'].value
C=np.ones([n,5])*np.asarray([x0[0],0,0,x0[1],0])
Co=8678.25/(sol_ac) #[mol/m3];
C=np.reshape(C.transpose(),5*n);

def ODE_2(t,C,ps):  
    k=np.asarray([ps['k1'].value,ps['k2'].value,ps['k3'].value,ps['k4'].value,ps['k5'].value,ps['k6'].value]);
    C=np.reshape(C,[5,n]).transpose();
    dC=np.empty([n,5])
    dC[:,0]=-2*k[0]*C[:,0]**2+k[2]*C[:,1]*C[:,3];
    dC[:,1]=k[0]*C[:,0]**2+k[1]*C[:,2]*C[:,4]-k[2]*C[:,1]*C[:,3]-k[4]*C[:,2]*C[:,1]-2*k[5]*C[:,1]**2;
    dC[:,2]=k[0]*C[:,0]**2-k[1]*C[:,2]*C[:,4]+k[2]*C[:,1]*C[:,3]-2*k[3]*C[:,2]**2-k[4]*C[:,2]*C[:,1];
    dC[:,3]=-k[2]*C[:,1]*C[:,3];
    
    dC[-1,-1]=D_ac*(2/(delta_r**2))*(C[-2,-1]-C[-1,-1])-k[1]*C[-1,-1]*C[-1,2];
    dC[1:-1,-1]=D_ac*(1/(delta_r**2))*(C[:-2,-1]-2*C[1:-1,-1]+C[2:,-1])-k[1]*C[1:-1,-1]*C[1:-1,2];
    dC[0,-1]=D_ac*(1/(delta_r**2))*(Co-2*C[0,-1]+C[1,-1])-k[1]*C[0,-1]*C[0,2];
    
    
    dC=np.reshape(dC.T,5*n);
    return dC 

t=np.linspace(0,3600*4,3600*4+1)
sol = solve_ivp(ODE_2,[0,t[-1]],C,method='BDF',args=(ps,),t_eval=t);
```

## Oxidacion En prueba de Headspace a Temp. Ambiente

```python
#Ecuacion Diferencial a solucionar
D_ac=1e-9 #[m2/s]
sol_ac=1.58E+03 #[Pa m3/mol]
A=0.0112 #[m2]
Vo=1e-4 #[m3]
P=101325 #[Pa]
P_N2=0.794*P#[Pa]

#Vector de radio discretizado
n=100
r=np.linspace(0,1.5e-4,n)
delta_r=r[1];


#constantes de reaccion
T=273+25
R=8.314
k=np.asarray([52691244*np.exp(-101.5e3/(R*T)),6581,572.68*np.exp(-31e3/(R*T)),2.2914e+08,10362358.9,96814085.92*np.exp(-48.4e3/(R*T))])

#Vector de inicializacion
ps=resultados_O2_80.params
Co=101325*0.206/(R*T) #[mol/m3];
C=sol.y[:,-1]
C[4*n]=Co

def ODE_2(t,C):  
    C=np.reshape(C,[5,n]).transpose();
    dC=np.empty([n,5])
    dC[:,0]=-2*k[0]*C[:,0]**2+k[2]*C[:,1]*C[:,3];
    dC[:,1]=k[0]*C[:,0]**2+k[1]*C[:,2]*C[:,4]-k[2]*C[:,1]*C[:,3]-k[4]*C[:,2]*C[:,1]-2*k[5]*C[:,1]**2;
    dC[:,2]=k[0]*C[:,0]**2-k[1]*C[:,2]*C[:,4]+k[2]*C[:,1]*C[:,3]-2*k[3]*C[:,2]**2-k[4]*C[:,2]*C[:,1];
    dC[:,3]=-k[2]*C[:,1]*C[:,3];
    
    V=(P_N2/(P-C[0,-1]*R*T))*Vo
    dC[-1,-1]=D_ac*(2/(delta_r**2))*(C[-2,-1]-C[-1,-1])-k[1]*C[-1,-1]*C[-1,2];
    dC[2:-1,-1]=D_ac*(1/(delta_r**2))*(C[1:-2,-1]-2*C[2:-1,-1]+C[3:,-1])-k[1]*C[2:-1,-1]*C[2:-1,2];
    dC[1,-1]=D_ac*(1/(delta_r**2))*((C[0,-1]*R*T/sol_ac)-2*C[1,-1]+C[2,-1])-k[1]*C[1,-1]*C[1,2];
    dC[0,-1]=D_ac*(1/(2*delta_r))*(-3*C[0,-1]*R*T/sol_ac +4*C[1,-1]-C[2,-1]) *(A/V);
    
    
    dC=np.reshape(dC.T,5*n);
    return dC 

t=np.linspace(3600*4,3600*24*15,3600+1)
sol_1= solve_ivp(ODE_2,[3600*4,t[-1]],C,method='BDF',t_eval=t);
```
